```python
columns_to_be_encoded=["Gender","Smoking","Hx Smoking","Hx Radiothreapy","Thyroid Function","Physical Examination","Adenopathy","Pathology","Focality","Risk","T","N","M","Stage","Response"]
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
one_hot_encoded = encoder.fit_transform(db[columns_to_be_encoded])
one_hot_db = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(columns_to_be_encoded))
db_encoded = pd.concat([db, one_hot_db], axis=1)
```

```python
db_encoded_1=db_encoded.drop(columns=columns_to_be_encoded)
```

```python
db_encoded_1.columns
db_encoded_1.head()
```

```python
db_encoded_2=db_encoded_1.drop(columns=["Gender_F","Smoking_No","Hx Smoking_No","Hx Radiothreapy_No"])
```

```python
db_encoded_2
```

```python
cols=["Recurred"]
one_hot_encoded = encoder.fit_transform(db_encoded_2[cols])
one_hot_db = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(cols))
db_encoded_3 = pd.concat([db_encoded_2, one_hot_db], axis=1)
```

```python
db_encoded_3.columns
```

```python
db_encoded_3=db_encoded_3.drop(columns=["Recurred","Recurred_No"])
```

```python
db_encoded_3.head()
```

```python
from sklearn.linear_model import LogisticRegression
my_model=LogisticRegression(random_state=42,max_iter=1000)
```

```python
from sklearn.model_selection import train_test_split
y=db_encoded_3["Recurred_Yes"]
X=db_encoded_3.drop(columns=["Recurred_Yes"])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
my_model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

# Make predictions on the test data
y_pred = my_model.predict(X_test)

# Calculate the accuracy score
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy*100}%")
```
